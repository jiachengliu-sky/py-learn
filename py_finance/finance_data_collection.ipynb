**Get interest rate from FRED**

````python
start = date(2000,1,1)
source = 'fred'
series_code = 'DGS10' # 10-year Treasure Rate

rates = DataReader(series_code, source, start)
rates
````

````python
# Get the index of max DGS10 for DGS10<3

rates.loc[rates.DGS10<3].idxmax()
````
