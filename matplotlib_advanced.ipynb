```python
fig, ax = plt.subplots()

seventies = climate_change["1970-01-01":"1979-12-31"]

ax.plot(seventies.index, seventies["co2"], c='b')
ax.set_ylabel("co2", c='b')
ax.tick_params('y', colors='b')

ax2 = ax.twinx()
ax2.plot(seventies.index, seventies["relative_temp"], c='r')
ax2.set_ylabel("Relative temperature", c='r')
ax2.tick_params('y', colors='r')

plt.show()
```

```python
# Helper function to plot with given variables
def plot_timeseries(axes, x, y, color, xlabel, ylabel):

  # Plot the inputs x,y in the provided color
  axes.plot(x, y, color=color)

  # Set the x-axis label
  axes.set_xlabel(xlabel)

  # Set the y-axis label
  axes.set_ylabel(ylabel, color=color)

  # Set the colors tick params for y-axis
  axes.tick_params('y', colors=color)
```

```python
# Plot using helper function plot_timeseries() defined above
fig, ax = plt.subplots()

plot_timeseries(ax, seventies.index, seventies["co2"], 
                'blue', "Time", "co2")
ax2=ax.twinx()
plot_timeseries(ax2, seventies.index, seventies["relative_temp"], 
                'red', "Time", "Relative Temp")

plt.show()
```

### Annotation

[matplotlib annotation api](https://matplotlib.org/users/annotations.html)

```python
fig, ax = plt.subplots()

plot_timeseries(ax, seventies.index, seventies["co2"], 
                'blue', "Time", "co2")
ax2=ax.twinx()
plot_timeseries(ax2, seventies.index, seventies["relative_temp"], 
                'red', "Time", "Relative Temp")

ax2.annotate(">0.2 degree", xy=(pd.Timestamp("1979-10-06"),0.2), 
             xytext=(pd.Timestamp('1979-01-06'),-0.2),
             arrowprops={"color":"gray", "arrowstyle":"->"})

plt.show()
```
